```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
set plotting limits
Xmin = -5.0; Xmax = 5.0; Npoints = 500
DelX = (Xmax-Xmin)/Npoints
x1 = np.arange(Xmin, Xmax, DelX)
x2 = np.arange(Xmin, Xmax, DelX/20)
y1 = -np.sin(x1) * np.cos(x1 * x1)
y2 = np.exp(-x2/4.) * np.sin(x2)

# plotting
plt.figure(1)
#2 rows, 1 column, 1st subplot
plt.subplot(2,1,1)
plt.plot(x1, y1, 'r', lw =2)
plt.xlabel('x'), plt.ylabel('f(x)'); plt.title('-sin(x)*cos(x^2)')
plt.grid(True)
#2 rows, 1 column, 2nd subplot
plt.subplot(2,1,2)
plt.plot(x2, y2, '-', lw =2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('exp(-x/4)*sin(x)')

# figure 2
plt.figure(2)
plt.subplot(2,1,1)
plt.plot(x1, y1*y1, 'r', lw =2)
plt.xlabel('x'), plt.ylabel('f(x)'); plt.title('sin^2(x)*cos^2(x^2)')
plt.subplot(2,1,2)
plt.plot(x2, y2*y2, '-', lw = 2)
plt.xlabel('x'), plt.ylabel('f(x)'); plt.title('exp(-x/2)*sin^2(x)')
plt.grid(True)

#show grpahs
plt.show()
```
