```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#title and labels
plt.title('Grade inflation')
plt.xlabel('Years in college')
plt.ylabel('GPA')
#for horizontal line
xa = np.array([-1,5])
ya = np.array([0,0])
#draw horizontal line
plt.plot(xa, ya)

#0-th Data set points
x0 = np.array([0,1,2,3,4])
y0 = np.array([-1.4,1.1,2.2,3.3,4.0])
#Draw points in blue
plt.plot(x0, y0, 'bo')
#draw line in green connecting points
plt.plot(x0, y0, 'g')

#1-st Data set points
x1 = np.arange(0, 5, 1)
y1 = np.array([4.0, 2.7, -1.8, -0.9, 2.6])
#draw line in red
plt.plot(x1, y1, 'r')

#asymmetric error bars
errTop = np.array([1.0, 0.3, 1.2, 0.4, 0.1])
errBot = np.array([2.0, 0.6, 2.3, 1.8, 0.4])
#plot error bars
plt.errorbar(x1, y1, [errBot, errTop], fmt ='o')

#grid line
plt.grid(True)
plt.show()
```
