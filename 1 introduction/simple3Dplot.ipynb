```python
import matplotlib.pylab as p
#need to import Axes3D from the matplotlib toolkit for 3D plotting
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
#create x and y arrays
delta = 0.1
x = p.arange(-3., 3., delta)
y = p.arange(-3., 3., delta)
#set up the entire coordinate matrix from the x and y vectors
X, Y = p.meshgrid(x, y)
Z = p.sin(X)*p.cos(Y) #construct the surface height

fig = p.figure() #create figure, the plot object
#ax: 3d axes object
ax = Axes3D(fig) #plot axes
ax.plot_surface(X, Y, Z) #surface
ax.plot_wireframe(X, Y, Z, color = 'r') #add wireframe
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

p.show()
```
